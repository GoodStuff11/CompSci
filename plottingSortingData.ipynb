```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename = ["sorting.csv","sortingReversed.csv","sortingSorted.csv"]
title = ["Completely Unorganized Arrays","Reversed Arrays","Ordered Arrays"]
for i in range(3):
    df = pd.read_csv(filename[i])
    
    plt.figure()
    plt.plot(df["Array size"],df["HeapSort comparisons"],label="Heap sort")
    plt.plot(df["Array size"],df["MergeSort comparisons"],label="Merge sort")
    plt.plot(df["Array size"],df["InsertionSort comparisons"],label="Insertion sort")
    plt.ylabel("Comparisons")
    plt.xlabel("Array Size")
    plt.title("Sorting " + title[i])
    plt.legend()

    plt.figure()
    plt.plot(df["Array size"],df["HeapSort comparisons"],label="Heap sort")
    plt.plot(df["Array size"],df["MergeSort comparisons"],label="Merge sort")
    plt.ylabel("Comparisons")
    plt.xlabel("Array Size")
    plt.title("Sorting " + title[i])
    plt.legend()
```
